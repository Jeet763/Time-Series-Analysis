## Time Resampling

```python
df_suzlon.head()
```

```python
df_suzlon.resample(rule='A').min()
```

```python
df_suzlon.resample(rule='A').max()
```

```python
##year end frequency
df_suzlon.resample(rule='A').max()['Open'].plot(figsize = (14,6))
```

```python
##quaterly start frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_suzlon.resample(rule='QS').max()['High'].plot()
```

```python
##Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_suzlon.resample(rule='BYE').max()
```

```python
df_suzlon.resample(rule='BQS').max()
```

```python
##plotting
df_suzlon['Open'].resample(rule='YE').mean().plot(kind='bar')
```

```python
df_suzlon['Open'].resample(rule='ME').max().plot(kind='bar',figsize=(15,6))
```
